**XGBoost - Classification**

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the dataset
train_data = pd.read_csv('/content/formatting_data.csv')
test_data = pd.read_excel('/content/test data.xlsx')

# Assuming your dataframe is called 'df' and threshold is a specific date
threshold = pd.to_datetime('2023-01-01')  # Replace with your threshold date

# Create output column: 'Yes' if date > threshold, 'No' otherwise
train_data['output'] = train_data['Transaction Date'].apply(lambda x: 'yes' if pd.to_datetime(x) > threshold else 'no')
test_data['output'] = test_data['Transaction Date'].apply(lambda x: 'yes' if pd.to_datetime(x) > threshold else 'no')

# 2. Separate features (X) and target (y) for training
X_train = train_data.drop('output', axis=1).select_dtypes(include=np.number)
y_train = train_data['output']

# 3. Separate features (X) and target (y) for testing
X_test = test_data.drop('output', axis=1).select_dtypes(include=np.number)
y_test = test_data['output']

# Ensure that X_train and X_test have the same columns
# Get the common columns between X_train and X_test
common_cols = list(set(X_train.columns) & set(X_test.columns))

# Select only the common columns in both X_train and X_test
X_train = X_train[common_cols]
X_test = X_test[common_cols]

# Convert 'yes' and 'no' to numerical labels (0 and 1)
y_train_numeric = y_train.map({'yes': 1, 'no': 0})
y_test_numeric = y_test.map({'yes': 1, 'no': 0})

# Convert the dataset into XGBoost's DMatrix format (optimized data structure)
dtrain = xgb.DMatrix(X_train, label=y_train_numeric)
dtest = xgb.DMatrix(X_test, label=y_test_numeric)

# Define XGBoost parameters
params = {
    'max_depth': 3,  # Maximum depth of a tree
    'eta': 0.1,      # Learning rate
    'objective': 'binary:logistic',  # Objective function for binary classification
    'eval_metric': 'logloss'  # Evaluation metric (log loss for binary classification)
}

# Train the model
num_rounds = 100  # Number of boosting rounds
model = xgb.train(params, dtrain, num_rounds)

# Make predictions on the test set
y_pred = model.predict(dtest)


# Convert predictions back to 'yes' and 'no' if needed
y_pred_labels = ['yes' if p >= 0.6 else 'no' for p in y_pred]

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_labels)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Get feature importances from the trained XGBoost model
feature_importance = model.get_score(importance_type='gain')

# Print detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_labels, zero_division=0))

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'Feature': list(feature_importance.keys()),
    'Importance': list(feature_importance.values())
})

# Print or display the feature importance
print("\nFeature Importance:")
print(feature_importance_df)
```
